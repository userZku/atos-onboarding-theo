### Question 1 — Compréhension de listes

Crée une fonction `pairs_au_carre(liste)` qui prend une liste de nombres et **retourne** une nouvelle liste contenant le carré des nombres pairs uniquement.

Exemple : `pairs_au_carre([1, 2, 3, 4, 5])` doit renvoyer `[4, 16]`.

````python
# Complète cette fonction
def pairs_au_carre(liste):
    # Ton code ici
    return [x**2 for x in liste if x % 2 == 0]


# --- Test (ne pas modifier) ---
SCORE["total"] += 1
try:
    assert pairs_au_carre([1, 2, 3, 4, 5]) == [4, 16]
    assert pairs_au_carre([0, -2, 7, 8]) == [0, 4, 64]
    assert pairs_au_carre([]) == []
    print("✅ Question 1 OK")
    SCORE["reussi"] += 1
except (AssertionError, TypeError) as e:
    print(f"❌ Question 1 — réessaie ({type(e).__name__})")
````

### Question 2 — Dictionnaires

Tu as une liste de dictionnaires représentant des transactions :

```python
transactions = [
    {"id": 1, "montant": 100, "type": "credit"},
    {"id": 2, "montant": 50, "type": "debit"},
    {"id": 3, "montant": 200, "type": "credit"},
]
```

Écris une fonction `total_credits(transactions)` qui retourne la **somme des montants** des transactions de type `"credit"`.

````python
# Complète cette fonction
def total_credits(transactions):
    # Ton code ici
    return sum(tx["montant"] for tx in transactions if tx["type"] == "credit")


# --- Test ---
SCORE["total"] += 1
try:
    txs = [
        {"id": 1, "montant": 100, "type": "credit"},
        {"id": 2, "montant": 50, "type": "debit"},
        {"id": 3, "montant": 200, "type": "credit"},
    ]
    assert total_credits(txs) == 300
    assert total_credits([]) == 0
    assert total_credits([{"id": 1, "montant": 99, "type": "debit"}]) == 0
    print("✅ Question 2 OK")
    SCORE["reussi"] += 1
except (AssertionError, TypeError) as e:
    print(f"❌ Question 2 — réessaie ({type(e).__name__})")
````

### Question 3 — Gestion d'exceptions

Écris une fonction `division_safe(a, b)` qui :
- retourne le résultat de `a / b` si la division est possible
- retourne la chaîne `"Division impossible"` si `b == 0`
- retourne la chaîne `"Type invalide"` si `a` ou `b` ne sont pas des nombres

Indice : `try`/`except` est ton ami.

````python
# Complète cette fonction
def division_safe(a, b):
    # Ton code ici
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        return "Type invalide"
    if b == 0:
        return "Division impossible"
    return a / b


# --- Test ---
SCORE["total"] += 1
try:
    assert division_safe(10, 2) == 5
    assert division_safe(7, 0) == "Division impossible"
    assert division_safe("dix", 2) == "Type invalide"
    print("✅ Question 3 OK")
    SCORE["reussi"] += 1
except (AssertionError, TypeError) as e:
    print(f"❌ Question 3 — réessaie ({type(e).__name__})")
````

### Question 9 — Filtrage de DataFrame

### Question 10 — Group by
