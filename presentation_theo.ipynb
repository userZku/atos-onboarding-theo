```python
# Un graphique simple commenté
# Ex pour Palmer Penguins : masse moyenne par espèce
df.groupby("species")["body_mass_g"].mean().plot.bar(
    title="Masse moyenne par espèce — Palmer Penguins"
)
plt.ylabel("Masse corporelle (g)")
plt.show()
```

**Observation (1 ligne)** : On distingue clairement une différence de masse moyenne entre certaines espèces de manchots.

### Voie B — Démo libre sur un dataset public non bateau
Dataset utilisé : Données de vins rouges (Wine Quality, UCI)

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
df_wine = pd.read_csv(url, sep=';')

# Afficher les 5 premières lignes
display(df_wine.head())

# Calculer la moyenne d'alcool par niveau de qualité
alcool_par_qualite = df_wine.groupby('quality')['alcohol'].mean()
print(alcool_par_qualite)

# Créer une nouvelle colonne binaire : bon_vin (qualité >= 7)
df_wine['bon_vin'] = df_wine['quality'] >= 7

# Afficher la répartition des bons vins
print(df_wine['bon_vin'].value_counts())

# Graphique : distribution de l'alcool selon la qualité
df_wine.boxplot(column='alcohol', by='quality', grid=False)
plt.title('Distribution du taux d\'alcool selon la qualité du vin')
plt.suptitle('')
plt.xlabel('Qualité')
plt.ylabel('Alcool (%)')
plt.show()
```

**Observation analytique :**
On observe que les vins de meilleure qualité ont tendance à avoir un taux d'alcool légèrement supérieur, mais la variabilité reste importante selon les lots.

## 5. Réponses aux 3 questions d'acculturation IA

### 5.1 Apprentissage supervisé vs non supervisé

En supervisé, on apprend à partir de données étiquetées pour prédire une cible connue. En non supervisé, il n’y a pas de cible: on cherche des structures cachées dans les données.

### 5.2 À quoi sert le transfer learning ?

Le transfer learning consiste à réutiliser un modèle déjà entraîné sur un gros corpus, puis à l’adapter à une tâche plus spécifique.

Exemple : un classifieur automatique de tickets.
On part d’un modèle de langage déjà pré-entraîné, puis on le réentraîne légèrement avec l’historique des tickets de l’entreprise (réseau, accès, logiciel, matériel).

### 5.3 Pourquoi un LLM n'est pas toujours la bonne réponse ?

Un LLM est puissant mais coûteux et imprécis sur des tâches simples.
Pour des règles métiers stables ou des calculs déterministes, un script classique est souvent plus fiable et plus facile à maintenir.

Par exemple utiliser un LLM pour valider un format de numéro client ou faire une conversion d’unités. Une simple méthode ferait l'affaire.
